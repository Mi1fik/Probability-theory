# Законы распределения дискретных случайных величин  
## Дополнительные примеры


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom, poisson, geom, hypergeom
```

## 1. Биномиальное распределение

### Пример задачи
Пусть X — число успешно выполненных заданий из 8,
если вероятность успешного выполнения одного задания равна 0.6.

Тогда:
$$
X \sim Bi(n=8, p=0.6)
$$


### Аналитический расчет

Закон распределения:
$$
P(X=k) = \binom{8}{k} 0.6^k 0.4^{8-k}
$$

Математическое ожидание:
$$
M[X] = np = 4.8
$$

Дисперсия:
$$
D[X] = np(1-p) = 1.92
$$

Среднее квадратическое отклонение:
$$
\sigma[X] \approx 1.386
$$

Мода:
$$
\lfloor (n+1)p \rfloor = 5
$$


```python
n, p = 8, 0.6
x = np.arange(0, n + 1)
pmf = binom.pmf(x, n, p)
mode_binom = x[np.argmax(pmf)]

print("M[X] =", binom.mean(n, p))
print("D[X] =", binom.var(n, p))
print("σ[X] =", binom.std(n, p))
print("мода =", mode_binom)

plt.stem(x, pmf, basefmt=' ')
plt.xlabel("x")
plt.ylabel("P(X)")
plt.title("Биномиальное распределение X ~ Bi(8, 0.6)")
plt.show()
```

## 2. Распределение Пуассона

### Пример задачи
Пусть Y — число звонков в службу поддержки за 10 минут,
если в среднем поступает 2 звонка.

Тогда:
$$
Y \sim П(\lambda = 2)
$$


### Аналитический расчет

Закон распределения:
$$
P(Y=k) = \frac{2^k e^{-2}}{k!}
$$

Математическое ожидание:
$$
M[Y] = 2
$$

Дисперсия:
$$
D[Y] = 2
$$

Среднее квадратическое отклонение:
$$
\sigma[Y] = \sqrt{2}
$$

$$
\text{мода} = 
\begin{cases}
\lfloor \lambda \rfloor, & \text{если } \lambda \text{ не целое} \\
\lambda, \lambda - 1, & \text{если } \lambda \text{ целое}
\end{cases}
$$
Мода: 2 или 1


```python
lam = 2
y = np.arange(0, 11)
pmf = poisson.pmf(y, lam)
max_prob = np.max(pmf)
mode_poisson = y[np.abs(pmf - max_prob) < 1e-10]

print("M[Y] =", poisson.mean(lam))
print("D[Y] =", poisson.var(lam))
print("σ[Y] =", poisson.std(lam))
print(f"Мода(ы): {list(mode_poisson)}")

plt.stem(y, pmf, basefmt=' ')
plt.xlabel("y")
plt.ylabel("P(Y)")
plt.title("Распределение Пуассона Y ~ Poisson(2)")
plt.show()
```

## 3. Геометрическое распределение

### Пример задачи
Пусть Z — число попыток до первого успешного подключения к серверу,
если вероятность успеха равна 0.4.

Тогда:
$$
Z \sim G(p=0.4)
$$


### Аналитический расчет

Закон распределения:
$$
P(Z=k) = (1-p)^{k-1} p
$$

Математическое ожидание:
$$
M[Z] = \frac{1}{p} = 2.5
$$

Дисперсия:
$$
D[Z] = \frac{1-p}{p^2} = 3.75
$$

Среднее квадратическое отклонение:
$$
\sigma[Z] \approx 1.936
$$

Мода: 1


```python
p = 0.4
z = np.arange(1, 11)
pmf = geom.pmf(z, p)
mode = z[np.argmax(pmf)]

print("M[Z] =", geom.mean(p))
print("D[Z] =", geom.var(p))
print("σ[Z] =", geom.std(p))
print(f"Мода = {mode}")


plt.stem(z, pmf, basefmt=' ')
plt.xlabel("z")
plt.ylabel("P(Z)")
plt.title("Геометрическое распределение Z ~ Geom(0.4)")
plt.show()
```

## 4. Гипергеометрическое распределение

### Пример задачи
Пусть W — число дефектных изделий среди 5 выбранных без возвращения
из партии из 12 изделий, где 4 дефектных.

Тогда:
$$
W \sim HG(N=12, M=4, n=5)
$$


### Аналитический расчет

Закон распределения:
$$
P(W = k) = \frac{\binom{4}{k}\binom{8}{5-k}}{\binom{12}{5}},
\quad k = 0,1,2,3,4
$$

Математическое ожидание:
$$
M[W] = n \cdot \frac{M}{N}
     = 5 \cdot \frac{4}{12}
     = \frac{20}{12}
     \approx 1.67
$$

Дисперсия:
$$
D[W] = n \cdot \frac{M}{N} \cdot \frac{N-M}{N} \cdot \frac{N-n}{N-1}
     \approx 0.707
$$

Среднее квадратическое отклонение:
$$
\sigma[W] \approx 0.841
$$

Мода:
$$
\operatorname{mode}(W) =
\left\lfloor \frac{(n+1)(M+1)}{N+2} \right\rfloor
= 2
$$


```python
N = 12
M = 4    
n = 5   

w = np.arange(0, n + 1)

pmf = hypergeom.pmf(w, N, M, n)

mean = hypergeom.mean(N, M, n)
var = hypergeom.var(N, M, n)
std = hypergeom.std(N, M, n)

mode = w[np.argmax(pmf)]

print(f"W: {w}")
print(f"P(W): {pmf}")
print(f"M[W] = {mean:.3f}")
print(f"D[W] = {var:.3f}")
print(f"σ[W] = {std:.3f}")
print(f"Мода = {mode}")
```

```python
plt.stem(w, pmf, basefmt=' ')
plt.xlabel("w")
plt.ylabel("P(W)")
plt.title("Гипергеометрическое распределение W ~ HG(12, 4, 5)")
plt.grid(True)
plt.show()
```
