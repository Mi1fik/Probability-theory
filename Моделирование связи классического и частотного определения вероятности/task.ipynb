## Постановка задачи

Рассмотрим эксперимент с одновременным бросанием двух честных
игральных кубиков.

Каждый кубик может принимать значения от 1 до 6,
все исходы равновероятны.

Событие A — сумма очков на двух кубиках равна 7.

Цель: показать, что при увеличении числа испытаний
частотная вероятность события A
стремится к классической вероятности.


## Классическая вероятность

Чётные числа при броске кости:
2, 4, 6 — всего 3 благоприятных исхода.

Общее число возможных исходов — 6.

Классическая вероятность события A:

$$
P(A) = \frac{3}{6} = \frac{1}{2} = 0.5
$$


```python
import random
import matplotlib.pyplot as plt
```

```python
def roll_die():
    return random.randint(1, 6)
```

```python
N_values = range(10, 5001, 50)
freq_probs = []

for N in N_values:
    count_even = 0
    for _ in range(N):
        if roll_die() % 2 == 0:
            count_even += 1
    freq_probs.append(count_even / N)
```

```python
classic_prob = 0.5
classic_probs = [classic_prob] * len(N_values)
```

```python
plt.figure(figsize=(10, 5))

plt.plot(N_values, freq_probs, label='Частотная вероятность')
plt.plot(
    N_values,
    classic_probs,
    linestyle='--',
    label='Классическая вероятность'
)

plt.xlabel('Число испытаний N')
plt.ylabel('Вероятность')
plt.title('Сходимость частотной вероятности к классической')
plt.legend()
plt.grid(True)

plt.show()
```

## Вывод

На графике видно, что при небольшом числе испытаний
частотная вероятность выпадения чётного числа
может значительно отклоняться от классической.

Однако с увеличением числа испытаний
частотная вероятность стремится к значению 0.5,
что подтверждает связь между частотным и
классическим определениями вероятности.
