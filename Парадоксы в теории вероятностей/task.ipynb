## Постановка задачи

Рассмотрим два метода лечения: A и B.

Пациенты разделены на две группы:
- лёгкое течение болезни
- тяжёлое течение болезни

В каждой группе метод A эффективнее метода B.

Покажем, что при объединении данных
метод B может оказаться эффективнее метода A.


## Исходные данные

Лёгкая форма заболевания:
- Метод A: 90 выздоровлений из 100
- Метод B: 80 выздоровлений из 100

Тяжёлая форма заболевания:
- Метод A: 10 выздоровлений из 100
- Метод B: 20 выздоровлений из 100

Метод A лучше в каждой группе.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {
    "A_light": (90, 100),
    "B_light": (80, 100),
    "A_heavy": (10, 100),
    "B_heavy": (20, 100)
}

p_A_light = data["A_light"][0] / data["A_light"][1]
p_B_light = data["B_light"][0] / data["B_light"][1]

p_A_heavy = data["A_heavy"][0] / data["A_heavy"][1]
p_B_heavy = data["B_heavy"][0] / data["B_heavy"][1]

p_A_light, p_B_light, p_A_heavy, p_B_heavy
```

```python
total_A = data["A_light"][1] + data["A_heavy"][1]
total_B = data["B_light"][1] + data["B_heavy"][1]

success_A = data["A_light"][0] + data["A_heavy"][0]
success_B = data["B_light"][0] + data["B_heavy"][0]

p_A_total = success_A / total_A
p_B_total = success_B / total_B

p_A_total, p_B_total
```

Метод B оказывается эффективнее в целом,
хотя метод A лучше в каждой подгруппе.


```python
labels = ['Лёгкая форма', 'Тяжёлая форма', 'Общая выборка']
A_values = [p_A_light, p_A_heavy, p_A_total]
B_values = [p_B_light, p_B_heavy, p_B_total]

x = np.arange(len(labels))
width = 0.35

plt.figure(figsize=(8, 5))
plt.bar(x - width/2, A_values, width, label='Метод A')
plt.bar(x + width/2, B_values, width, label='Метод B')

plt.xticks(x, labels)
plt.ylabel('Вероятность выздоровления')
plt.title('Парадокс Симпсона')
plt.legend()
plt.grid(axis='y')

plt.show()
```

## Объяснение парадокса

Парадокс возникает из-за **разного распределения пациентов**
по подгруппам:

- метод A чаще применялся в тяжёлых случаях
- метод B — в лёгких

При объединении данных вклад подгрупп
искажает общую картину.


## Вывод

Парадокс Симпсона показывает, что:

- анализ только агрегированных данных может быть ошибочным
- необходимо учитывать структуру выборки
- вероятности в подгруппах и в целом могут вести себя противоположно

Это подчёркивает важность корректной интерпретации
вероятностных и статистических данных.
